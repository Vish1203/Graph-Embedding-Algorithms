```python
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score, f1_score

import os
import networkx as nx
import numpy as np
import pandas as pd
import time

from stellargraph.data.explorer import BiasedRandomWalk
from stellargraph import StellarGraph
from stellargraph import datasets
from IPython.display import display, HTML

%matplotlib inline
```

```python
data = datasets.Cora()
display(HTML(data.description))
g, node_sub = data.load(largest_connected_component_only=True)
```

```python
print(g.info())
```

```python
first = time.localtime(time.time())
print(first)
random_walk = BiasedRandomWalk(g) #Node2VEc
walks = random_walk.run(
nodes=list(g.nodes()),  # root nodes
    length=50,  # maximum length of a random walk
    n=10, #No. of random walks per root node
    p=0.5, #Probability, 1/p of returning to source node 
    q=2.0) #Probability, 1/q of moving away from source node
print("Number of random walks: {}".format(len(walks)))
last = time.localtime(time.time())
print(last)
#print(last - first)
```

```python
from gensim.models import Word2Vec

str_walks = [[str(n) for n in walk] for walk in walks]
model = Word2Vec(str_walks, size=128, window=5, min_count=0, sg=1, workers=4)
```

```python
model.wv["1107010"].shape
```

```python
node_ids = model.wv.index2word  # list of node IDs
node_embeddings = (
    model.wv.vectors
)  # numpy.ndarray of size number of nodes times embeddings dimensionality
node_targets = node_sub[[int(node_id) for node_id in node_ids]]
```

```python
tsne = TSNE(n_components=2)
node_embeddings_2d = tsne.fit_transform(node_embeddings)
```

```python
label_map = {l: i for i, l in enumerate(np.unique(node_targets))}
node_colours = [label_map[target] for target in node_targets]

plt.figure(figsize=(10, 8))
plt.scatter(
    node_embeddings_2d[:, 0],
    node_embeddings_2d[:, 1],
    c=node_colours,
    cmap="jet"
)
```

```python
#X will hold the 128-dimensional input features
X = node_embeddings
# y holds the corresponding target values
y = np.array(node_targets)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.1, test_size=None)
```

```python
print(
    "Array shapes:\n X_train = {}\n y_train = {}\n X_test = {}\n y_test = {}".format(
        X_train.shape, y_train.shape, X_test.shape, y_test.shape
    )
)
```

```python
clf = LogisticRegressionCV(
    Cs=10, cv=10, scoring="accuracy", verbose=False, multi_class="ovr", max_iter=300
)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='micro')
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
import seaborn as sns

Mul_cm = metrics.confusion_matrix(y_test, y_pred)
plt.show()
plt.figure()
sns.heatmap(Mul_cm);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
plt.title('Confusion Matrix');

report = classification_report(y_test, y_pred)
print(report)
              
```
